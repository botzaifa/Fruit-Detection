```python
import cv2 
import numpy as np

# Read image
image = cv2.imread("apple.jpg") 
# Convert BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Color strength parameters in HSV for red color (adjust as needed)
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

# Threshold HSV image to obtain red color
mask = cv2.inRange(hsv, lower_red, upper_red) 

# Show original image and result
cv2.imshow('Image', image)
cv2.imshow('Result', mask) 

# Press any key to exit
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np

# Read image
image = cv2.imread("apple2.jpg") 
# Convert BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Color strength parameters in HSV for red color (adjust as needed)
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

# Threshold HSV image to obtain red color
mask = cv2.inRange(hsv, lower_red, upper_red) 

# Find contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw bounding boxes around detected objects (assuming the largest contour is the fruit)
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(image, 'Fruit', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save the labeled image
cv2.imwrite("labeled_apples.jpg", image)

# Show original image and result
cv2.imshow('Image', image)
cv2.imshow('Result', mask) 

# Press any key to exit
cv2.waitKey(0)
cv2.destroyAllWindows()
```
